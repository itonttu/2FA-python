```python
import qrcode
import pyotp
import time

random_base32 = pyotp.random_base32()

uri = pyotp.totp.TOTP(random_base32).provisioning_uri(name="itotube1234@gmail.com", issuer_name="サンプルアプリ")

img = qrcode.make(uri)
img.save('qr_code.png')

totp = pyotp.TOTP(random_base32)
for i in range(1000):
    print(totp.now())
    time.sleep(1)
```
